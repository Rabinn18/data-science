```python
def univariate_barplot(df, title, xlabel, ylabel, figsize, colors):
    plt.figure(figsize= figsize)
    columns = list(df.columns)
    ax = sns.barplot(x=columns[0],y=columns[1],data = df, palette = colors)
    plt.title("Distribution ofselling price in dataset")
    plt.xlabel('x label')
    plt.ylabel('y label')
    for value in ax.patches:
        ax.text(value.get_x(), value.get_height() + .5, value.get_height())
    plt.show()
```

### Is Winning a toss a decider IN IPl Games

```python
pd.crosstab(top8_ipl_teams['TossWinner'], top8_ipl_teams['WinningTeam'])
```

```python
top8_ipl_teams.columns
```

```python
toss_decider = pd.crosstab(top8_ipl_teams['TossWinner'],
                             top8_ipl_teams['WinningTeam'], margins = True)
toss_decider.head()
```

```python
#Null Hypothesis
import scipy.stats as stats
alpha = 0.05
Ho = "H0: There is no relationship between winning Toos and winning game"
H1= "H1: winning a toss is an advantage in winningteam2"
test_stats = stats.chi2_contingency(observed = toss_decider)
p_value = test_stats[1]
if p_value <= alpha:    
    print(H1)
else:
    print(H2)
```

```python
league_matches = top8_ipl_teams.loc[~top8_ipl_teams['MatchNumber'].isin(removed_match)]
league_matches
```

```python
home_decider = pd.crosstab(league_matches['Team1'],
                             league_matches['WinningTeam'], margins = True)
home_decider
```

```python
#Null Hypothesis
alpha = 0.05
Ho = "H0: Team does not win in home ground"
#Alternative Hypothesis
H1= "H1: Team wins in home ground"
test_stats = stats.chi2_contingency(observed = home_decider)
p_value = test_stats[1]
if p_value <= alpha:    
    print(H1)
else:
    print(H2)
```
