```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = open('auto-mpg.data', 'r')
fileread = file.readlines()
```

```python
print(fileread)
```

```python
data_of_data = []
for item in fileread:
    data = []
    item = item.replace("\n" , "")
    item = item.replace("\t" , " ")
    item = item.split(' ')
    for subitem in item:
        if subitem != "":
            data.append(subitem)
    data_of_data.append(data)
#     print(item)
```

```python
new_data_of_data = []
for item in data_of_data: 
#     print(len(item))
    new_data = item[:8]
    item = ' '.join(subitem for subitem in item[8:])
    new_data.append(item)
    new_data_of_data.append(new_data)
```

```python
print(new_data_of_data)
```

```python
file = open('auto-mpg.names', 'r')
fileread = file.read()
# fileread = fileread.split('\n')
```

```python
fileread
```

```python
column_values = fileread.split('Attribute Information:')[1]
column_values = column_values.split('\n')
column_values
```

```python
columns = []
for col in column_values[2:-4]:
    strip_col = col.strip()
    split_col = strip_col.split(':')[0].split('.')[1].strip()
    print(split_col)
    columns.append(split_col)
```

```python
print(columns)
```

```python
cars_df = pd.DataFrame(data = new_data_of_data, columns = columns)
cars_df.head()
```

```python
cars_df.to_csv('cars_df.csv', index = False)
```

```python
cars_df.columns = ["mpg", "cyl", "disp", "hp", "wt",
                 "acc", "year", "org", "name"]
```

## Creating Regression Model for disp and mpg

```python
clean_cars_df.info()
```

```python
final_df = clean_cars_df[['mpg', 'disp', 'hp','wt']]
```

```python
final_df.info()
```

```python
final_df = final_df.astype('float64')
final_df
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(data = final_df, x = 'disp', y = 'mpg')
plt.title('Displacement VS Miles per Gallon')
plt.show()
```

```python
sns.kdeplot
```

```python
data = final_df[['disp']]
labels = final_df['mpg']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test= train_test_split(data, labels, test_size = 0.25, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, y_pred, squared = False)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(data = final_df, x = 'disp', y = 'mpg',color='violet')
# sns.scatterplot(data = final_df, x = 'hp', y = 'mpg',color='purple')
plt.plot(x_test, y_pred, color='indigo')
plt.title('Displacement VS Miles per Gallon')
plt.show()
```
