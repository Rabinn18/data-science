```python
## speed comparision between python list and numpy array
```

```python
import time
```

```python
first_list =  [i for i in range (1, 10000000)]
```

```python
print(type(first_list))
```

```python
first_ndarray = np.array(first_list)
print(type(first_ndarray))
```

```python
start_time = time.perf_counter()
sum_first_list = sum(first_list)
end_time = time.perf_counter()
list_processing_time = end_time - start_time
print(list_processing_time)
```

```python
start_time = time.perf_counter()
sum_first_list = np.sum(first_ndarray)
end_time = time.perf_counter()
ndarray_processing_time = end_time - start_time
print(ndarray_processing_time)
```

```python
print("numpy array is {} times faster than python list".format(list_processing_time/ndarray_processing_time))
```

# Iterating NumPy Arrays

```python
import numpy as np
```

```python
arr = np.array([1,3,5,6,7])
for item in arr:
    print(item)
```

```python
arr2d = np.random.randint(1,25,(4,3))
for item in arr2d:
    for subitem in item:
        print(subitem)
```

```python
arr3d = np.random.randint(1,25,(3,3,3))
for item in arr3d:
    for subitem in item:
        for individualitem in subitem:
            print(individualitem)
```

```python
#using np.nditer() to iterate elements of array one by one
```

```python
for item in np.nditer(arr):
    print(item)
```

```python
for item in np.nditer(arr2d):
    print(item)
```

```python
for item in np.nditer(arr3d):
    print(item)
```

```python
#using np.ndenumerate()
```

```python
for idx, item in np.ndenumerate(arr):
    print(idx, item)
```

```python
for idx, item in np.ndenumerate(arr2d):
    print(idx, item)
```

```python
for idx, item in np.ndenumerate(arr3d):
    print(idx, item)
```

# Copy Vs View In Numpy Arrays:

```python
arr1 = np.random.randint(1,25,(9))
arr2 = np.random.randint(25,50,(9))
```

```python
print(arr1)
print(arr2)
```

```python
arr3 = np.copy(arr1)
print(arr3)
```

```python
arr4 = arr1.view()
print(arr4)
```

```python
# changes made in view array will effect in original array and vice versa
# original array is immune to changes made in copy array vice versa
```

```python
arr3[1]=20
```

```python
print(arr1)
```

```python
arr4[1]=20
```

```python
print(arr1)
```

```python
arr1[1]=14
```

```python
print(arr3)
```

```python
print(arr4)
```

```python
#now for 2dd array
```

```python
print(arr5)
```

```python
arr6 = np.copy(arr5)
print(arr6)
```

Types of Stacking
There are three types of stacking. They are as follows:
    -> Stacking along rows:NumPy provides a helper function: hstack() to stack along rows.
    -> Stcaking along columns:To stack two arrays along columns, Numpy also provides a helper function: vstack().
    -> Stacking along depth:Another helper function to stack two numpy arrays along their depth, we have dstack().

```python
arr2d1 = np.random.randint(1,20,(2,2))
print(arr2d1)
arr2d1.shape
```

```python
arr2d2 = np.random.randint(20,30,(2,2))
print(arr2d2)
arr2d2.shape
```

```python
hstack2d = np.hstack([arr2d1, arr2d2])
print(hstack2d)
hstack2d.shape
```

```python
#similar to concatenating with axis =1
np.concatenate([arr2d1, arr2d2], axis =1)
```

```python
vstack2d = np.vstack([arr2d1, arr2d2])      #similar to concatenating with axis =1
print(vstack2d)
vstack2d.shape
```

```python
dstack2d = np.dstack([arr2d1, arr2d2])
print(dstack2d)
dstack2d.shape
```
